```python
#import library
import numpy as np  
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#read data
df_can = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)
```

```python
df_can.head()
```

```python
df_can.info()
```

```python
#drop kolom yang tidak dibutuhkan
# in pandas axis=0 represents rows (default) and axis=1 represents columns.
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head(5)
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.columns
```

```python
df_can['Total'] = df_can.sum(axis=1)
```

**Question 1**: Let's compare the number of immigrants from India and China from 1980 to 2013

```python
#no 1
df_can.set_index('Country', inplace=True)
Tahun = list(range(1980,2014))
df_ic = df_can.loc[['India', 'China'], Tahun].transpose()
df_ic
```

```python
df_ic.plot(figsize=(25,5))
plt.title('Imigran China dan India (1980 - 2013) ')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Imigran')
```

**Question 2**: Compare the trend of top 5 countries that contributed the most to immigration to Canada.

```python
#no 2
df_can.sort_values(by='Total', ascending=False, inplace=True)
df_5 = df_can.head(5)
df_5 = df_5[Tahun]
df_5
```

```python
df_5 = df_5.transpose()
df_5.plot(kind='area',stacked=False,figsize=(25,5))
plt.title(' top 5 countries that contributed the most to immigration to Canada')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Imigran')
```

**Question 3**: Create an unstacked area plot of the 5 countries that contributed the least to immigration to Canada from 1980 to 2013. Use a transparency value of 0.55.

```python
#no 3
df_b5 = df_can.tail(5)
df_b5 = df_b5[Tahun].transpose()
df_b5.plot(kind='area',stacked=False, alpha=0.55, figsize=(18,8))
plt.title(' 5 countries that contributed the least to immigration to Canada')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Imigran')
```

**Question 4**: Display the immigration distribution for Greece, Albania, and Bulgaria for years 1980 - 2013? Use an overlapping plot with 15 bins and a transparency value of 0.35.

```python
#no 4
df_gralbu = df_can.loc[['Greece', 'Albania', 'Bulgaria'], Tahun].transpose()
count, bin_edge = np.histogram(df_gralbu, 15)
df_gralbu.plot(kind='hist', bins=15, xticks= bin_edge, stacked=False, alpha=0.35, figsize=(18,8))
plt.title('  Immigration distribution for Greece, Albania, and Bulgaria for years 1980 - 2013')
plt.xlabel('Tahun')

plt.ylabel('Jumlah Imigran')
```

**Question 5**: Create a horizontal bar plot showing the total number of immigrants to Canada from the top 15 countries, for the period 1980 - 2013. Label each country with the total immigrant count.

```python
#no 5
df_15 = df_can['Total'].head(15)
df_15.values
```

```python
df_15.plot(kind='barh',figsize=(20,10))

for index, value in enumerate(df_15):
  label = format(int(value), ',')
  plt.annotate(label, xy=(value - 47000, index - 0.1), color='white')
```
